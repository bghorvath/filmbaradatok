```python
from pyannote.audio import Pipeline
```

```python
with open("huggingface.token","r") as f:
    ACCESS_TOKEN = f.read().strip()
```

```python
pipeline = Pipeline.from_pretrained("pyannote/speaker-segmentation", use_auth_token=ACCESS_TOKEN)
```

```python
pipeline2 = Pipeline.from_pretrained("pyannote/speaker-diarization", use_auth_token=ACCESS_TOKEN)
```

```python
# apply pretrained pipeline
segmentation = pipeline(f"240_sampled.wav")
```

```python
diarization = pipeline2(f"240_sampled.wav") 
```

```python
segmentation
```

```python
diarization
```

```python
import pickle

with open("diarization.pkl", "wb") as f:
    pickle.dump(diarization, f)
```

```python
# print the result
for i, (turn, _, speaker) in enumerate(segmentation.itertracks(yield_label=True)):
    print(f"start={turn.start:.1f}s stop={turn.end:.1f}s speaker_{speaker}")
```

```python
for i, (turn, _, speaker) in enumerate(diarization.itertracks(yield_label=True)):
    print(f"start={turn.start:.1f}s stop={turn.end:.1f}s speaker_{speaker}")
```
