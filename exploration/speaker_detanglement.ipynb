### Merge same speaker

```python
diary = []
start = None
stop = None
speaker = None
for i, segment in enumerate(diarization):
    new_start = segment["start"]
    new_stop = segment["stop"]
    new_speaker = segment["speaker"]

    # first segment
    if speaker is None:
        start = new_start
        stop = new_stop
        speaker = new_speaker

    # if same speaker, append segment
    if new_speaker == speaker and new_start - stop < 10:
        stop = new_stop
        if i == len(diarization):
            diary.append(
                {
                    "start": start,
                    "stop": stop,
                    "speaker": speaker,
                }
            )
    
    # else, save last one, create new one
    else:
        diary.append(
            {
                "start": start,
                "stop": stop,
                "speaker": speaker,
            }
        )
        start = new_start
        stop = new_stop
        speaker = new_speaker
```
